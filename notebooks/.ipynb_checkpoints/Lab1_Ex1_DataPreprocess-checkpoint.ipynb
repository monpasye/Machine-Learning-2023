* `hip`
* `hop`
* `hip-hop`

Чтобы избавиться от них, создадим функцию ***replace_wrong_genres()*** с двумя параметрами:
* `wrong_genres=` — список дупликатов
* `correct_genre=` — строка с правильным значением

Функция должна исправить имена в столбце 'genre' из таблицы df, т.е. заменить каждое значение из списка `wrong_genres` на значение в `correct_genre`.

```python
def replace_wrong_genres (wrong_genres, correct_genre):
    for wrong_genre in wrong_genres:
        df['genre'] = df['genre'].replace(wrong_genre, correct_genre)
```

```python
duplicates = ['hip','hop','hip-hop']
name = 'hiphop'
replace_wrong_genres(duplicates, name)  # удаление неявных дубликатов
```

Создайте функцию ***number_tracks()***, чтобы подсчитать количество воспроизведенных песен в определенный день и город. Для этого потребуются два параметра:
* день недели
* название города

В функции используйте переменную для хранения строк из исходной таблицы, где:
   * Значение столбца `'day'` равно параметру `'day`
   * Значение столбца `'city'` равно параметру `'city``

Примените последовательную фильтрацию с логической индексацией.

Затем вычислите значения столбца 'user_id' в результирующей таблице. Сохраните результат в новую переменную. Верните эту переменную из функции.

```python
def number_tracks (day_of_the_week,name_of_the_city):
    day= day_of_the_week
    city = name_of_the_city

# Пусть переменная track_list сохранит строки таблицы df, где
# значение в столбце 'day' равно параметру day и в то же время
# значение в столбце 'city' равно параметру city (применить последовательную фильтрацию
# с логической индексацией).
    track_list = df.loc[(df['day'] == day)&(df['city'] == city),['user_id','day','city']]

# Пусть переменная track_list_count хранит количество значений столбца user_id в track_list
# (находится с помощью метода count()).
    track_list_count = 0
    track_list_count += track_list['user_id'].count()
    return track_list_count

# Пусть функция возвращает число: значение track_list_count.
# Функция подсчитывает прослушанные треки для определенного города и дня.
    return track_list_count

# Сначала он извлекает из таблицы строки с предполагаемым днем,
# затем отфильтровывает из результата строки с предполагаемым городом,
# затем находит количество значений в столбце 'user_id' в отфильтрованной таблице,
# затем возвращает это число.
# Чтобы увидеть, что она возвращает, выполните вызов функции в print().
    print(track_list_count)
```

Вызов функции `number_tracks()` шесть раз, изменив значения параметров, чтобы получить данные по обоим городам за каждый из трех дней.

```python
# количество песен, сыгранных в Спрингфилде в понедельник
number_tracks ('Monday','Springfield')
```

```python
# количество песен, сыгранных в Спрингфилде в среду
number_tracks ('Wednesday','Springfield')
```

```python
# количество песен, сыгранных в Шелбивилле в среду
number_tracks ('Wednesday','Shelbyville')
```

```python
# количество песен, сыгранных в Спрингфилде в пятницу
number_tracks ('Friday','Springfield')
```

```python
# количество песен, сыгранных в Шелбивилле в пятницу
number_tracks ('Friday','Shelbyville')
```

Напишите функцию ***genre_weekday()*** с четырьмя параметрами:
* Таблица данных (`df`)
* День недели («day»)
* Первая временная метка в формате «чч: мм» («time1»)
* Последняя временная метка в формате «чч:мм» (`time2`)

Функция должна возвращать информацию о 15 самых популярных жанрах в определенный день в течение периода между двумя временными метками.

```python
# 1) Пусть в переменной жанра_df будут храниться строки, соответствующие нескольким условиям:
# - значение в столбце day равно значению аргумента day=
# - значение в столбце time больше значения аргумента time1=
# - значение в столбце time меньше значения аргумента time2=
# Используйте последовательную фильтрацию с логической индексацией.

# 2) Сгруппировать жанр_df по столбцу «жанр», взять один из его столбцов,
# и используйте метод count(), чтобы найти количество записей для каждого из
# представленные жанры; сохраните полученную серию в
# переменная жанра_df_count

# 3) Отсортируйте жанр_df_count в порядке убывания частоты и сохраните результат
# в переменную жанра_df_sorted

# 4) Вернуть объект Series с первыми 15 значениями жанра жанра_df_sorted — 15 наиболее
# популярных жанров (в определенный день, в течение определенного периода времени)

def genre_weekday(df,day,time1,time2):
     # последовательная фильтрация
     # Создайте переменную жанр_df, которая будет хранить только те строки df, где день равен day=
    genre_df = df.loc[df['day'] == day]

    # снова фильтруем, чтобы жанр_df сохранял только те строки, в которых время меньше, чем time2=
    genre_df1 = genre_df.loc[genre_df['time'] < time2]

    # фильтруем еще раз, чтобы жанр_df сохранял только те строки, где время больше, чем time1=
    genre_df2 = genre_df1.loc[(genre_df1['time'] > time1)]

    # группируем отфильтрованный DataFrame по столбцу с названиями жанров, берем столбец жанра и методом count() находим количество строк для каждого жанра
    genre_df_count = genre_df2.groupby ('genre')['time'].count()

    # сортируем результат по убыванию (чтобы в объекте Series на первом месте стояли самые популярные жанры)
    genre_df_sorted = genre_df_count.sort_values(ascending=False)

    # мы вернем объект Series, хранящий 15 самых популярных жанров в определенный день в заданный период времени
    return genre_df_sorted[:15]
```

Сравните результаты функции genre_weekday() для Спрингфилда и Шелбивилля в понедельник утром (с 7:00 до 11:00) и в пятницу вечером (с 17:00 до 23:00):

```python
# вызов функции для утра понедельника в Спрингфилде
genre_weekday(spr_general,'Monday','07:00','11:00')
```

```python
# вызов функции для утра понедельника в Шелбивилле
genre_weekday(shel_general,'Monday','07:00','11:00')
```

```python
# вызов функции для вечера пятницы в Спрингфилде
genre_weekday(spr_general,'Friday','17:00','23:00')
```

```python
# вызов функции на вечер пятницы в Шелбивилле
genre_weekday(shel_general,'Friday','17:00','23:00')
```

**Выводы**

Сравнив топ-15 жанров в понедельник утром, можно сделать следующие выводы:

# *Гипотеза 3: У слушателей Спрингфилда и Шелбивилля разные предпочтения. В Спрингфилде предпочитают музыку в стиле "поп", а в Шелбивилле больше поклонников рэпа.* <a id='genre'></a>


Сгруппируйте таблицу spr_general по жанрам и найдите количество воспроизведенных песен для каждого жанра с помощью метода count(). Затем отсортируйте результат в порядке убывания и сохраните его в spr_genres.

```python
# в одну строку: сгруппируйте таблицу spr_general по столбцу «жанр»,
# подсчитайте значения жанра с помощью count() в группировке,
# сортируем полученную серию в порядке убывания и сохраняем ее в spr_genres

spr_genres = spr_general.groupby('genre')['track'].count().sort_values(ascending=False)
```
